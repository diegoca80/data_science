```python
from numpy import array,dot,count_nonzero
```

```python
import plotly as py
```

```python
unit_step = lambda x: 0 if x < 0 else 1
```

```python
with open("dados_lista_1_2016/dados1.txt","r") as f:
    data = f.readlines()
    x = []
    for d,l in zip(data[101:],data[:100]):
        a,b,c = d.split(";")
        x.append([float(a),float(b),float(c),unit_step(float(l))])
```

```python
alfa = 0.2
w = [0,0,0]
```

```python
while True:
        errors = []
        for features in x:
            #scalar product
            result = dot(w,array(features[:3]))
            error = features[3] - unit_step(result)
            errors.append(error)
            w += alfa * error * array(features[:3])
        if(count_nonzero(errors) == 0):
            print(errors)
            print(w)
            break
```
